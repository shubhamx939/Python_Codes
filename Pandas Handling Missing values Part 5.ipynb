# Handling Missing values of a file in Python (Part 5):

```python
import pandas as pd
```

```python
df = pd.read_csv("test.csv")
df
```

# Syntax ======>>>> df.replace(to_replace=None, value=None, inplace=False, limit=None, regex=False, method='pad',)

```python
#replacing desired value into new value by giving parameter ==>> to_replace="value which needs to be changed" & value="new value":

df.replace(to_replace = "standard", value = "classic")
```

```python
#Here 1st parameter is "to_replace" and 2nd parameter is "value": Same as above 

df.replace("some college", "HKBKCE")
```

```python
#replacing integer values:

df.replace(88,88888)
```

```python
#Passing a list as a parameter in df.replace([list]):

df.replace([72,74],[7200,7400])
```

```python
#Passing dictionary as the parameter of df.replace({dict}):

#We are using Regular Expression to catch the values of lunch column and then replacing it with 9999 
#where regex=True means we are confirming that we are explicitly using Regular Expression in this statement: 

df.replace({"lunch": "[A-Za-z]"}, 9999, regex=True)
```
