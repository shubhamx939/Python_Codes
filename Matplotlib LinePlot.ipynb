```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("delhiTemp.csv")
df
```

```python
x = df["Date"]
x
```

```python
y = df["Temp"]
y
```

```python
#Putting title(), xlabel() & ylabel() :

plt.plot(x,y)

plt.title("New Delhi Temperature Graph")
plt.xlabel("Date")
plt.ylabel("Temperatue")

plt.show()
```

```python
#Putting custom axis, color, marker, linestyle, linewidth, markersize :

plt.plot(x,y, color="g", marker="*", linestyle="--", linewidth=2, markersize=10)

plt.title("New Delhi Temperature Graph")
plt.xlabel("Date")
plt.ylabel("Temperatue")

plt.axis([0,35, 0,30])

plt.show()
```

```python
#Putting label, color & fontsize to title-xlabel-ylabel, legend() :

plt.plot(x,y, "g*--", linewidth=2, markersize=10, label="Temp")

plt.title("New Delhi Temperature Graph", color="Red", fontsize=15)
plt.xlabel("Date",  color="black", fontsize=15)
plt.ylabel("Temperatue",  color="black", fontsize=15)

plt.axis([0,35, 0,30])

plt.legend(loc="lower right")

plt.show()
```

```python
#using style for ggplot or grids :

from matplotlib import style

style.use("ggplot")

plt.plot(x,y, "m*--", linewidth=2, markersize=10, label="Temp")

plt.title("New Delhi Temperature Graph", color="Red", fontsize=15)
plt.xlabel("Date",  color="black", fontsize=15)
plt.ylabel("Temperatue",  color="black", fontsize=15)

plt.axis([0,35, 0,30])
plt.legend(loc="lower right")

plt.show()
```
