```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(21,45,(100))
py_student_age = np.random.randint(15,40,(100))
```

```python
ml_student_age
```

```python
py_student_age
```

```python
plt.hist(ml_student_age)
plt.title("ML Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins)
plt.title("ML Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.8)
plt.title("ML Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.8, orientation="horizontal")
plt.title("ML Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.8, orientation="horizontal", color="grey")
plt.title("ML Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.8, orientation="horizontal", color="grey", label="ML Student Age")
plt.title("ML Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.legend(loc="lower right")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.8, orientation="horizontal", color="grey", label="ML Student")

plt.hist(py_student_age, bins, rwidth=0.8, orientation="horizontal", color="m", label="PY Student")

plt.title("ML & PY Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.legend(loc="lower right")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.figure(figsize=[16,9])
plt.hist([ml_student_age,py_student_age], bins, rwidth=0.5, orientation="vertical", color=["grey","orange"], label=["ML Student","PY Student"])

plt.title("ML & PY Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.legend(loc="upper left")
plt.show()
```

```python
from matplotlib import style

style.use("ggplot")

bins=[15,20,25,30,35,40,45]

plt.figure(figsize=[16,9])
plt.hist([ml_student_age,py_student_age], bins, rwidth=0.5, orientation="vertical", color=["grey","orange"], label=["ML Student","PY Student"])

plt.title("ML & PY Student Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("No of Students")
plt.legend(loc="upper left")
plt.show()
```
