# Printing specific columns from a csv file

```python
df = pd.read_csv("Fortune_10.csv", usecols = [0])
df
```

```python
df = pd.read_csv("Fortune_10.csv", usecols = [0,1])
df
```

```python
df = pd.read_csv("Fortune_10.csv", usecols = [0,2,4,7])
df
```

# Skipping Rows of a CSV File (Single row & Multiple rows)

```python
df = pd.read_csv("Fortune_10.csv")
df
```

```python
df = pd.read_csv("Fortune_10.csv", skiprows = 1)
df
```

```python
df = pd.read_csv("Fortune_10.csv", skiprows = 2)
df
```

```python
df = pd.read_csv("Fortune_10.csv", skiprows = [0])
df
```

```python
df = pd.read_csv("Fortune_10.csv", skiprows = [2])
df
```

```python
df = pd.read_csv("Fortune_10.csv", skiprows = [0,2,3])
df
```

# Creating specific column as the ID

```python
df = pd.read_csv("Fortune_10.csv", index_col = "ID")
df
```

```python
df = pd.read_csv("Fortune_10.csv", index_col = "Name")
df
```

```python
df = pd.read_csv("Fortune_10.csv", index_col = 4)
df
```
