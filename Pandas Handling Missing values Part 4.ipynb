# Handling Missing values of a file in Python (Part 4):

```python
import pandas as pd
```

```python
df = pd.read_csv("test.csv")
df
```

### Syntax====>>> df.fillna(value=None,method=None,axis=None,inplace=False,limit=None,downcast=None,)

```python
df.fillna(99999)
```

```python
#Passing dictionary as a parameter in "fillna(dict)":

df.fillna({"math score": 8000})
```

```python
#Filling "previous_value" of same column to the "NaN Values":

df.fillna(method = "ffill")
```

```python
#Filling "next_value" of same column to the "NaN Values":

df.fillna(method = "bfill")
```

```python
#Filling "previous_value" of same column to the "NaN Values":

#Same as method = "ffill"


df.fillna(method = "pad")
```
