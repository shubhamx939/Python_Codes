```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python","R", "AI", "ML", "DS"]

class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.barh(classes,class1_students)
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5)
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="edge")
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="center", color="grey")
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="edge", color="orange", edgecolor="black", linewidth=3)
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="edge", color="orange", edgecolor="black", linewidth=3, linestyle="--")
plt.title("Classes Bar Chart")
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="edge", color="orange", edgecolor="black", linewidth=3, linestyle="-", label="Class1 Students")
plt.title("Classes Bar Chart")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=[16,9])
plt.bar(classes,class1_students, width=0.5, align="center", color="orange", edgecolor="black", linewidth=3, linestyle="-", label="Class1 Students")
plt.title("Classes Bar Chart")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.barh(classes,class1_students, align="center", color="orange", edgecolor="black", linewidth=3, linestyle="-", label="Class1 Students")
plt.title("Classes Bar Chart")
plt.legend()
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="center", color="orange", edgecolor="black", linewidth=3, linestyle="-", label="Class1 Students")
plt.title("Classes Bar Chart", fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No of Students",fontsize=15)
plt.legend()
plt.show()
```

```python
plt.bar(classes,class1_students, width=0.5, align="center", color="orange", edgecolor="black", linewidth=3, linestyle="-", label="Class1 Students")
plt.bar(classes,class2_students, width=0.5, align="center", color="red", edgecolor="black", linewidth=3, linestyle="-", label="Class2 Students")


plt.title("Classes Bar Chart", fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No of Students",fontsize=15)
plt.legend()
plt.show()
```

```python
class_index = np.arange(len(classes))
width = 0.2

plt.figure(figsize=[16,9])
plt.bar(class_index,class1_students , width, color="grey", label="Class1 Students")
plt.bar(class_index + width,class2_students , width, color="blue", label="Class2 Students")
plt.bar(class_index + width + width,class3_students , width, color="orange", label="Class3 Students")

plt.xticks(class_index + width, classes, rotation=30)

plt.title("Classes Bar Chart", fontsize=28)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No of Students",fontsize=25)
plt.legend()
plt.show()
```

```python
class_index = np.arange(len(classes))
width = 0.2

plt.figure(figsize=[16,9])
plt.barh(class_index,class1_students , width, color="grey", label="Class1 Students")
plt.barh(class_index + width,class2_students , width, color="blue", label="Class2 Students")
plt.barh(class_index + width + width,class3_students , width, color="orange", label="Class3 Students")

plt.yticks(class_index + width, classes, rotation=30)

plt.title("Classes Bar Chart", fontsize=28)
plt.ylabel("Classes",fontsize=25)
plt.xlabel("No of Students",fontsize=25)
plt.legend()
plt.show()
```
