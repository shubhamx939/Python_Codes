```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("googleplaystore.csv")
df
```

```python
df.isnull().sum()
```

```python
df1 = pd.read_csv("googleplaystore.csv", nrows = 1000)
```

```python
df1
```

```python
x = df1["Rating"]
y = df1["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.scatter(x,y, color="red")
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y, color="red")
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y, color="red", marker="*")
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y, color="red", marker="*", s=100)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y, color="red", marker="*", s=100, alpha=0.5)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y, color="red", marker="*", s=100, alpha=0.5, linewidth=10)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y, color="red", marker="*", s=100, alpha=0.5, linewidth=10, edgecolor="green")
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```
