# Changing the Data Type of a column using "dtype" argument in "read_csv()" method:

```python
df = pd.read_csv("Fortune_10.csv", dtype = {"ID":float})
df
```

```python
df = pd.read_csv("Fortune_10.csv", dtype = {"ID": float, "Profit":float})
df
```

# Changing values of columns into "TRUE" 7 "FALSE" using "true_values" & "false_values" argument of "read_csv()" method:

```python
df = pd.read_csv("Fortune_10.csv")
df
```

```python
df = pd.read_csv("Fortune_10.csv", true_values = ["Yes"], false_values = ["No"])
df
```
