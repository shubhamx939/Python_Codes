```python
import matplotlib.pyplot as plt
```

```python
classes = ["PYTHON","AI", "ML", "DL", "NLP","DJANGO"]

class_students = [70,55,62,34,26,65]
```

```python
plt.pie(class_students)
```

```python
plt.pie(class_students, labels = classes)
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]

plt.pie(class_students, labels = classes, explode = explode_values)
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]
color_values = ["c","b","r","m","g","y"]

plt.pie(class_students, labels = classes, explode = explode_values, colors = color_values)
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]
color_values = ["c","b","r","m","g","y"]

plt.pie(class_students, labels = classes, explode = explode_values, colors = color_values, autopct = "%0.2f%%")
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]
color_values = ["c","b","r","m","g","y"]

plt.pie(class_students, labels = classes, explode = explode_values, colors = color_values, autopct = "%0.2f%%", radius = 2.4)
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]
color_values = ["c","b","r","m","g","y"]

plt.pie(class_students, labels = classes, explode = explode_values, 
        colors = color_values, autopct = "%0.2f%%", radius = 2.4, startangle=90)
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]
color_values = ["c","b","r","m","g","y"]
font_size = {"fontsize":18}

plt.pie(class_students, labels = classes, explode = explode_values, 
        colors = color_values, autopct = "%0.2f%%", radius = 2.4, startangle=90, textprops=font_size)
plt.show()
```

```python
explode_values = [0,0,0.2,0,0,0]
color_values = ["c","b","r","m","g","y"]
font_size = {"fontsize":18}

plt.pie(class_students, labels = classes, explode = explode_values, 
        colors = color_values, autopct = "%0.2f%%", radius = 2.4, startangle=90, textprops=font_size, shadow=True)
plt.show()
```
